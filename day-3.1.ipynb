If-elif-else statement

```python
a = 0

if a > 0:
    print("a is a positive number")
elif a < 0:
    print("a is a negative number")
else:
    print("a is zero")
```

User input

```python
num1 = float(input("Enter number 1 : "))
num2 = float(input("Enter number 2 : "))
num3 = float(input("Enter number 3 : "))

if num1 >= num2 and num1 >= num3:
    print("num1 : ", num1, " is the largest number")
elif num2 >= num1 and num2 >= num3:
    print("num2 : ", num2, " is the largest number")
else:
    print("num3 : ", num3, " is the largest number")
```

Nested if

```python
num1 = float(input("Enter number 1 : "))
num2 = float(input("Enter number 2 : "))
num3 = float(input("Enter number 3 : "))

if num1 >= num2 and num1 >= num3:
    if num2 >= num3:
        print("Descending order : ", num1, ", ", num2, ", ", num3)
    else:
        print("Descending order : ", num1, ", ", num3, ", ", num2)
elif num2 >= num1 and num2 >= num3:
    if num1 >= num3:
        print("Descending order : ", num2, ", ", num1, ", ", num3)
    else:
        print("Descending order : ", num2, ", ", num3, ", ", num1)
else:
    if num1 >= num2:
        print("Descending order : ", num3, ", ", num1, ", ", num2)
    else:
        print("Descending order : ", num3, ", ", num2, ", ", num1)
```
