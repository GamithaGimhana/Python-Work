If-elif-else statement

```python
a = 0

if a > 0:
    print("a is a positive number")
elif a < 0:
    print("a is a negative number")
else:
    print("a is zero")
```

User input

```python
num1 = float(input("Enter number 1 : "))
num2 = float(input("Enter number 2 : "))
num3 = float(input("Enter number 3 : "))

if num1 >= num2 and num1 >= num3:
    print("num1 : ", num1, " is the largest number")
elif num2 >= num1 and num2 >= num3:
    print("num2 : ", num2, " is the largest number")
else:
    print("num3 : ", num3, " is the largest number")
```

Nested if

```python
num1 = float(input("Enter number 1 : "))
num2 = float(input("Enter number 2 : "))
num3 = float(input("Enter number 3 : "))

if num1 >= num2 and num1 >= num3:
    if num2 >= num3:
        print("Descending order : ", num1, ", ", num2, ", ", num3)
    else:
        print("Descending order : ", num1, ", ", num3, ", ", num2)
elif num2 >= num1 and num2 >= num3:
    if num1 >= num3:
        print("Descending order : ", num2, ", ", num1, ", ", num3)
    else:
        print("Descending order : ", num2, ", ", num3, ", ", num1)
else:
    if num1 >= num2:
        print("Descending order : ", num3, ", ", num1, ", ", num2)
    else:
        print("Descending order : ", num3, ", ", num2, ", ", num1)
```

Find Absolute value

```python
num = float(input("Enter a number : "))

# abs_value = abs(num)    # Using built-in function

abs_value = num if num >= 0 else -num
print("Absolute value of ", num, " is ", abs_value)
```

Structural pattern matching

```python
x = 2

match x:           # pattern matching (Python 3.10+)
    case 0:
        print("x is zero")
    case 1:
        print("x is one")
    case 2:
        print("x is two")
    case _:
        print("x is something else")
```

Activity with or

```python
day = input("Enter a day : ")

match day:
    case "monday" | "tuesday" | "wednesday" | "thursday" | "friday":    # | means OR, mek methnadi bitwise or ek kiyl ganan gnne na, normal or ekk widiyt wada krnne
        print("Weekday")
    case "saturday" | "sunday":
        print("Weekend")
    case _:
        print("Invalid day")
```

Matching by the length of an Iterable

```python
numbers = [1, 2, 3]

match numbers:
    case [x, y]:
        print(x * y)
    case [x, y, z]: 
        print(x + y + z)
    case _:
        print("Nothing matched")
```

Activity

```python
triangle_sides = [3, 4, 5]

match triangle_sides:
    case [x, y]:  
        print(x ** 2 + y ** 2 ** 0.5)
    case [x, y, z]:
        a = (x + y + z) / 2
        area = (a * (a - x) * (a - y) * (a - z)) ** 0.5
        print("Area of triangle is : ", area)
    case _:
        print("Unsupported number of input")
```

for loop

```python
nums = [5, 10, 4, 9, 1]

for item in nums:
    print(item)

print("Loop is finished")
```
