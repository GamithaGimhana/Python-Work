Activity get sum of squares

```python
def sum_of_squares(n):
    if n < 1:
        return "Give a positive number"

    sum = 0
    for i in range(1, n+1):
        sum += i**2 
    return sum

print(sum_of_squares(2))
```

Activity add corresponding numbers

```python
def add_corresponding_numbers(list1, list2): 
    if len(list1) != len(list2):
        return "Error: Lists are not of equal length"
    
    corresponding_list = []
    for i in range(len(list1)):
        corresponding_list.append(list1[i] + list2[i])

    return corresponding_list

list1 = [1,2,3,4,5]
list2 = [1,2,3,4,5]

print(add_corresponding_numbers(list1, list2))
```

Activity

```python
def calculate_total(price, qty = 1, discount = 0):
    total = price * qty
    total -= total * (discount / 100)
    return total

print(calculate_total(100))
print(calculate_total(100, 2))
print(calculate_total(100, 2, 10))
```

Activity

```python
def sum_numbers(*args):
    sum = 0

    for num in args:
        sum += num

    return sum

print(sum_numbers(1,2,3,4,5))
print(sum_numbers(10,20,30))
```

Activity

```python
def average_positive(*args):
    sum = 0
    count = 0

    for num in args:
        if num > 0:
            sum += num
            count += 1

    if count == 0:
        return "No positive numbers are provided."
    
    return sum / count

print(average_positive(1, -2, 3, -4, 5))
print(average_positive(-1, -2, -3))
print(average_positive(10, 20))
```

Arbitrary Keyword Arguments

```python
def arbitrary_keyword_args(**kwargs):
    print(kwargs)
    print(type(kwargs))

arbitrary_keyword_args(name="Gamitha", age=21, city="Meegoda")
arbitrary_keyword_args(a=1, b=2, c=3)
```

Example

```python
def arbitrary_keyword_args(**kwargs):
    for key in kwargs.keys():
        print(key)

    for value in kwargs.values():
        print(value)

    for key, value in kwargs.items():
        print(key + ": " + str(value))

arbitrary_keyword_args(name="Gamitha", age= 21, city="Meegoda")
arbitrary_keyword_args(a=1, b=2, c=3)
```

Activity find average score

```python
def average_score(**subjects):    
    sum = 0
    count = 0

    # if not subjects:
    #     return "No subjects are provided."

    for score in subjects.values():
        sum += score
        count += 1

    if count == 0:
        return "No subjects are provided."

    return sum / count

print(average_score(Math=90, Science=80, English=85))
print(average_score())
```

without built-in

```python
def absolute_value(x):
    if x < 0:
        return -x
    
    return x

print(absolute_value(-5))
print(absolute_value(3))
```

with built-in

```python
print(abs(-5))
print(abs(3))
```

map()

```python
def find_square(x):
    return x ** 2

my_list = [1,2,3,4,5]

result = map(find_square, my_list)
print(result)                   # <map object at 0x0000025C1484B490>
print(type(result))             # <class 'map'>
                                        
result_list = list(result)  
print(result_list)              # [1, 4, 9, 16, 25]
print(type(result_list))        # <class 'list'>
```

Activity Fahrenheit to Celsius

```python
def fahrenheit_to_celsius(f_temperatures):
    def convert_to_celsius(f):
        celsius = (f - 32) * 5/9
        return round(celsius, 2)    # Rounding to 2 decimal places
    
    c_temperatures = list(map(convert_to_celsius, f_temperatures))
    print(c_temperatures)

fahrenheit_to_celsius([32, 68, 100, 212])
```

filter()

```python
def even(num):
    return num % 2 == 0

list1 = [9, 5, 8, 12, 13]

even_numbers = list(filter(even, list1))
print(even_numbers) 
```

Activity

```python
def devide(num):
    return num % 3 == 0 and num % 5 != 0
    
numbers = [12, 15, 18, 20, 25, 30, 35]

result = list(filter(devide, numbers))
print(result) 
```
