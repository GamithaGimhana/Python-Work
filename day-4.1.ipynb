Functions and Functional Programming

Activity function define

```python
def sum(a,b):
    sum_result = a + b
    return sum_result

print(sum(3,5))
```

Activity get sum of squares

```python
def sum_of_squares(n):
    if n < 1:
        return "Give a positive number"

    sum = 0
    for i in range(1, n+1):
        sum += i**2 
    return sum

print(sum_of_squares(2))
```

Activity add corresponding numbers

```python
def add_corresponding_numbers(list1, list2): 
    if len(list1) != len(list2):
        return "Error: Lists are not of equal length"
    
    corresponding_list = []
    for i in range(len(list1)):
        corresponding_list.append(list1[i] + list2[i])

    return corresponding_list

list1 = [1,2,3,4,5]
list2 = [1,2,3,4,5]

print(add_corresponding_numbers(list1, list2))
```

```python
def add(a, b=10):
    return a+b

print(add(3,4)) 
print(add(3))
# print(add(3,4,5))  # This will raise a TypeError
```

Activity

```python
def calculate_total(price, qty = 1, discount = 0):
    total = price * qty
    total -= total * (discount / 100)
    return total

print(calculate_total(100))
print(calculate_total(100, 2))
print(calculate_total(100, 2, 10))
```

Keyword Arguments

```python
def add(a,b):
    return a + b

print(add(a=5, b=7))
print(add(b=7, a=5))
```

Arbitrary Positional Arguments

```python
def arbitrary_positional_args(*args):
    print(args)
    print(type(args))

arbitrary_positional_args(1,2,3,4,5)
arbitrary_positional_args('a','b','c') 
arbitrary_positional_args(1,2,3)
```

Activity

```python
def sum_numbers(*args):
    sum = 0

    for num in args:
        sum += num

    return sum

print(sum_numbers(1,2,3,4,5))
print(sum_numbers(10,20,30))
```
