isinstance()

```python
x = 10

print(isinstance(x, int))  # False
print(isinstance(x, float))  # True
print(isinstance(x, (int, float)))  # True
```

Object Oriented Programming

Classes

```python
class Parrot:
    name = ""
    age = 0

parrot_1 = Parrot()

print(parrot_1.name)
print(parrot_1.age)
```

Methods

```python
class Player:
    score = 10

    def add_points(self, score): # self kiyn ek hama method ekktm pass wenn oni. parameter ekk thibune nathath ek dann oni
        self.score += score

player_1 = Player()
player_1.add_points(5)
print(player_1.score)
```

Activity

```python
class Car:
    wheels = 4

    def drive_to(self, destination):
        print(f"Driving to {destination} with {self.wheels} wheels.")

    def show_wheel_count(self):
        print(f"This car has {self.wheels} wheels.")

my_car = Car()
my_car.drive_to("Colombo")
my_car.show_wheel_count()
```
