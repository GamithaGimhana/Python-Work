Activity extended with constructor

```python
class Car:
    wheels = 4

    def __init__(self, model, color):
        self.model = model
        self.color = color

    def drive_to(self, destination):
        print(f"The {self.color} {self.model} is driving to {destination}.")

    def show_wheel_count(self):
        print(f"This {self.color} {self.model} runs on {self.wheels} wheels.")

my_car = Car("BMW", "black")
my_car.drive_to("Colombo")
my_car.show_wheel_count()
```

Activity

```python
from math import pi

class Circle:
    pi = pi

    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return pi * (self.radius ** 2)
    
    def circumference(self):
        return 2 * pi * self.radius
    
    def display_info(self):
        print(f"Circle with radius: {self.radius}")
        print(f"Area: {round(self.area(), 2)}")
        print(f"Circumference: {round(self.circumference(), 2)}")

circle_1 = Circle(10)
circle_1.display_info()
```

Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} speaks.")

class Dog(Animal):
    pass    

my_dog = Dog("Nova")
my_dog.speak()
```

Override parent class method from child class

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} speaks.")

class Dog(Animal):
    def speak(self):
        print(f"{self.name} barks.")    

my_dog = Dog("Nova")
my_dog.speak()
```

Override parent class constructor from child class

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print(f"{self.name} speaks.")

class Dog(Animal):
    def __init__(self, name, color):
        super().__init__(name)
        self.color = color
        
    def speak(self):
        print(f"{self.color} {self.name} barks.")    

my_dog = Dog("Nova", "Black")
my_dog.speak()
```
