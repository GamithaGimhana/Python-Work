Polymorphism

```python
class Polygon:
    # method to render a shape
    def render(self):
        print("Rendering Polygon...")

class Square(Polygon):
    # renders Square
    def render(self):
        print("Rendering Square...")

class Circle(Polygon):
    # renders circle
    def render(self):
        print("Rendering Circle...")

# create an object of Square
s1 = Square()
s1.render()

# create an object of Circle
c1 = Circle()
c1.render()
```

Encapsulation

protected

```python
class Person:
    def __init__(self):
        self._name = "Gamitha"

class Employee(Person):
    def display(self):
        print("Employee Name:", self._name)

e = Employee()
print(e._name)  # Attempting to access the protected member directly
```

Private

```python
class Dog:
    def __init__(self, name):
        self.__name = name  # Private attribute
    def get_name(self):
        return self.__name

dog_1 = Dog("Buddy")
print(dog_1.get_name())  # Accessing the private member via a public method

# print(dog_1.__name)  # Attempting to access the private member directly
# This will raise an AttributeError because __name is a private attribute
```

```python
class Person:
    def __init__(self):
        self.__name = "Gamitha"

class Employee(Person):
    def display(self):
        print("Employee Name:", self._Person__name)

e = Employee()
print(e._Person__name)  # Attempting to access the protected member directly
```

Getter & Setter methods

```python
class BankAccount:
    def __init__(self):
        self.__balance = 0 # Private attribute

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient funds")

account = BankAccount()

account.deposit(1000)
print(account.get_balance()) # Accessing balance using getter

account.withdraw(500)
print(account.get_balance()) # Accessing balance using getter
```

Activity

```python
class StudentRecord:
    def __init__(self):      
        self.__grades = []

    def get_grades(self):
        return self.__grades
    
    def add_grade(self, mark):
        if self.__validate_grade(mark):
            self.__grades.append(mark)
        else:
            print("Grade must be an integer between 0 and 100.")

    def __validate_grade(self, mark):
        return 0 <= mark <= 100 and isinstance(mark, int)
        
    def get_average(self):
        return self.__calculate_average()

    def __calculate_average(self):
        if not self.__grades:
            return 0
        return sum(self.__grades) / len(self.__grades)

student = StudentRecord()

student.add_grade(75)
student.add_grade(80)
student.add_grade(95)

print("Average Grade: ", student.get_average())
```

Example

```python
class BankAccount:
    def __init__(self, account_number, initial_balance):
        self.__account_number = account_number
        self.__balance = initial_balance

    def __validate_amount(self, amount):
        return amount > 0

    def __update_balance(self, amount):
        if self.__validate_amount(amount):
            self.__balance += amount

    def __check_withdrawal(self, amount):
        return self.__validate_amount(amount) and amount <= self.__balance

    def deposit(self, amount):
        self.__update_balance(amount)

    def withdraw(self, amount):
        if self.__check_withdrawal(amount):
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

    def get_account_number(self):
        return self.__account_number

# Example usage:
account1 = BankAccount("123456", 1000.0)
print(f"Account Number: {account1.get_account_number()}")
print(f"Initial Balance: ${account1.get_balance()}")
account1.deposit(500.0)
#account1.__validate_amount(500.0) #This will generate an error
print(f"Balance after deposit: ${account1.get_balance()}")
account1.withdraw(200.0)
print(f"Balance after withdrawal: ${account1.get_balance()}")
```
