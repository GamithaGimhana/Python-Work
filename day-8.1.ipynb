Check response code

```python
import requests

url = "https://api.github.com/events"
 
response = requests.get(url)

if response.status_code == 200:
    print("Fetching Successful.")
elif response.status_code == 404:
    print("Fetching Failed - Resource not found.")
elif response.status_code == 403:
    print("Fetching Failed - Access forbidden.")
elif response.status_code == 500:
    print("Fetching Failed - Internal Server error.")
else:
    print("Fetching Failed - Received unexpected status code:", response.status_code)
```

response json

```python
import requests

url = "https://api.github.com/events"
 
response = requests.get(url)


if response.headers['Content-Type'] == 'application/json; charset=utf-8':
    print("Response is in JSON format.")
    print(response.json())
    print(type(response.json()))
    print(response.headers['Content-Type'])
else:
    print("Response is not in JSON format.")
```

Print first 100 characters using slicing

```python
import requests

url = "https://api.github.com/events"
 
response = requests.get(url)

if response.status_code == 200:
    print("Fetching Successful.")
    print(response.text[:100])  # Print first 100 characters using slicing
else:
    print("Fetching Failed: ", response.status_code)
```

post

```python
import requests

data = {
    "username": "user",
    "password": "pass"
}

response = requests.post("https://httpbin.org/post", data)
if response.status_code == 200:
    print("POST request successful.")
    print(response.text)
else:
    print("POST request failed with status code:", response.status_code)
```

url parameters

```python
# url = 'https://jsonplaceholder.typicode.com/?userId=1&id=2'

import requests

url = 'https://jsonplaceholder.typicode.com/posts/'
params = {
    'userId': 1,
    'id': 2
}

response = requests.get(url, params=params)
if response.status_code == 200:
    print("GET request successful.")
    print(response.url)  # Full URL with query parameters
    print(response.text)
else:
    print("GET request failed with status code:", response.status_code)
```

Basic Authentication

```python
import requests

url = "https://httpbin.org/get"

username = 'user'
password = 'pass'

response = requests.get(url, auth=(username, password)) # Basic Authentication - mekedi auth parameter ekata tuple ekak danna oni (username, password).
if response.status_code == 200:
    print("Fetching Successful.")
    print(response.text)
else:
    print("Fetching Failed with status code:", response.status_code)
```

Beautiful Soup Library

```python
from bs4 import BeautifulSoup

html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""

soup = BeautifulSoup(html_doc, 'html.parser')

print(soup.p)           # Prints the first <p> tag - mek attribute ekk widiyt denne
print(soup.a)           # Prints the first <a> tag - mek attribute ekk widiyt denne

print(soup.find('p'))   # Finds and prints the first <p> tag - mek method ekk widiyt denne
print(soup.find('a'))   # Finds and prints the first <a> tag - mek method ekk widiyt denne

print(soup.find_all('a'))  # Finds and prints all <a> tags in a list
print(soup.find_all('p'))  # Finds and prints all <p> tags in a list

print(soup.title)       # Prints the <title> tag
print(soup.title.text)  # Prints the text inside the <title> tag
```

Activity - use of Request and BeautifulSoup libraries

```python
import requests
from bs4 import BeautifulSoup

url = "https://example.com"

response = requests.get(url)

if response.status_code == 200:
    print("Fetching Successful.")
    # print("Response Content:", response.text)
    # print("Content Type:", response.headers['Content-Type'])
    soup = BeautifulSoup(response.text, 'html.parser')

    h1_tag = soup.h1.text

    if h1_tag:
        print("H1 Tag Content:", h1_tag)
    else:
        print("H1 Tag not found.")
else:
    print("Fetching Failed with status code:", response.status_code)
```
