File Handling

file opening

```python
file_1 = open("my_file.txt", "r")

content = file_1.read()
print(content)

file_1.close()
```

file opening using with 

```python
with open("my_file.txt", "r") as file_1:
    content = file_1.read()
    print(content)
```

readline method

```python
with open("my_file.txt", "r") as file_1:
    line_1 = file_1.readline()
    print(line_1)
    line_2 = file_1.readline()
    print(line_2)
    line_3 = file_1.readline()
    print(line_3)
    
```

strip method

```python
with open("my_file.txt", "r") as file_1:
    line_1 = file_1.readline()
    print(line_1.strip())
    line_2 = file_1.readline()
    print(line_2.strip())
    line_3 = file_1.readline()
    print(line_3.strip())
    
```

read lines of a file using one readline method

```python
with open("my_file.txt", "r") as file_1:
    while True:
        line = file_1.readline()
        if not line:
            break
        print(line.strip())
```

readlines method

```python
with open("my_file.txt", "r") as file_1:
    lines = file_1.readlines()
    print(lines)
```

file creation

```python
with open("new_file.txt", "w") as file_1:
    file_1.write("Hello, World!\n")
    file_1.write("This is a new file.\n")
```

override a file

```python
with open("new_file.txt", "w") as file_1:
    file_1.write("Hello, I'm Gami!\n")
    file_1.write("Working in a new file.\n")
```

writelines method

```python
with open("new_file.txt", "w") as file_1:
    lines = ["Line 1\n", "Line 2\n", "Line 3\n"]
    file_1.writelines(lines)
```

append files

```python
with open("new_file.txt", "a") as file_1:
    file_1.write("Appending a new line.\n")
```

file creation with append

```python
with open("append_file.txt", "a") as file_1:    # file ekk nathnm e namin aluthin file ekk create krnw
    file_1.write("Appending a new line.\n")
```

json module

```python
import json

with open("my_json.json", "r") as json_file:
    data = json.load(json_file)  # json file eke data tika python dict akakata convert krnw
    print(data)
    print(type(data))
```
